```python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Definimos la función para graficar las cónicas

def conicas(a, b, c, d, e, f):
    '''
    Esta función grafica las cónicas en 3D.
    '''
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)

    Z = a*X**2 + b*X*Y + c*Y**2 + d*X + e*Y + f

    ax.contour3D(X, Y, Z, 50, cmap='viridis')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
```

```python
conicas(1, 2, 1, 0, 0, 0)
```

```python
# Ahora vamos a hacer una funcio para graficar una cuadrica

def plot_quadric(a, b, c, d, e, f):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)

    Z = (a * X**2) + (b * Y**2) + (c * X * Y) + (d * X) + (e * Y) + f

    ax.plot_surface(X, Y, Z, cmap='viridis')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
```

```python
plot_quadric(1, 2, 0, 0, 0, 0)
```
